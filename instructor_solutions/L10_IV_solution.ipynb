**Example (Distance to Surgery Center):**
1. **Relevance:** Strong. People who live closer are more likely to get surgery.
2. **Exclusion:** Fragile. Living closer to a center might also mean better access to rehab, specialists, or a healthier environment, which affects the outcome directly.
3. **Independence:** Fragile. People with higher socioeconomic status might choose to live in areas with better medical facilities, and SES is a major cause of health outcomes ($U$).

## 1. Setup

```python
import pandas as pd
import numpy as np
from phs564_ci.datasets import load_data
from phs564_ci.estimators.iv import wald_estimator, iv_2sls

df = load_data("l10_iv.csv")
```

## 2. Wald and 2SLS Estimation

```python
wald_late = wald_estimator(df, 'Y', 'A', 'Z')
print(f"Wald LATE: {wald_late:.3f}")

iv_res = iv_2sls(df, 'Y', 'A', 'Z', [])
print(f"2SLS Estimate: {iv_res.params['A']:.3f}")
```
