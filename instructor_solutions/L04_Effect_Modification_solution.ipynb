## ðŸ›‘ Activity 1: Choose scale for 2 questions
**Answer 1:** The **Risk Difference (RD)**. If Group V=1 has an RD of 0.30 and Group V=0 has an RD of 0.10, you save 30 people per 100 treated in Group 1 vs. only 10 in Group 0. 

**Answer 2:** Often the **Risk Ratio (RR)**. If the RR is 2.0 in both groups, the drug doubles the risk relative to baseline, which might suggest a consistent mechanism even if the absolute risk reduction differs.

## 1. Setup

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from phs564_ci.datasets import load_data

df = load_data("l04_effect_modification.csv")
```

## 2. Effect Measures by Stratum

```python
risks = df.groupby(['V', 'A'])['Y'].mean().unstack()
risks['RD'] = risks[1] - risks[0]
risks['RR'] = risks[1] / risks[0]
print(risks[['RD', 'RR']])
```

## 3. Regression Interaction

```python
model = smf.ols("Y ~ A * V", data=df).fit()
print(model.params)
```

**Note:** The `A:V` coefficient represents the difference in RDs ($RD_{V=1} - RD_{V=0}$).
