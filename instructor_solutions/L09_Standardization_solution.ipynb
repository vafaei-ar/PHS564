## ðŸ›‘ Activity 1: Implement g-formula (toy)
**Step-by-step check:**
1. Model fit: `smf.logit("Y ~ A + L", data=df).fit()`
2. Counterfactuals: Everyone is A=1, then everyone is A=0.
3. Prediction: Use `model.predict(df_counterfactual)`.
4. ATE: `predictions_a1.mean() - predictions_a0.mean()`.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from phs564_ci.datasets import load_data

df = load_data("l09_standardization.csv")
model = smf.logit("Y ~ A + L", data=df).fit(disp=0)

df_a1 = df.copy(); df_a1['A'] = 1
df_a0 = df.copy(); df_a0['A'] = 0

rd = model.predict(df_a1).mean() - model.predict(df_a0).mean()
print(f"Standardized RD: {rd:.3f}")
```
