# Lecture 06 Solution: Confounding and Adjustment Strategies

## ðŸ›‘ Activity 1: Choose adjustment set
**Answers:**
1. **Confounder:** `L1` (common cause of A and Y).
2. **Mediator:** `M` (on the path A -> M -> Y).
3. **Collider:** `L2` (common effect of A and Y).
4. **Adjustment Set:** `{L1}` only. Adjusting for `M` would block the effect; adjusting for `L2` would induce selection bias.

## 1. Setup

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from phs564_ci.datasets import load_data
from phs564_ci.diagnostics.balance import calculate_smd

df = load_data("l06_confounding.csv")
```

## 2. Regression vs. Matching

```python
model = smf.ols("Y ~ A + L1 + L2", data=df).fit()
print(f"Regression RD: {model.params['A']:.3f}")

from sklearn.neighbors import NearestNeighbors
treated = df[df['A'] == 1]
control = df[df['A'] == 0]
nn = NearestNeighbors(n_neighbors=1).fit(control[['L1', 'L2']])
indices = nn.kneighbors(treated[['L1', 'L2']])[1]
matched_control = control.iloc[indices.flatten()]
matched_rd = treated['Y'].mean() - matched_control['Y'].mean()
print(f"Matched RD: {matched_rd:.3f}")
```
