# Lecture 11 Solution: Inverse Probability Treatment Weighting (IPTW)

## ðŸ›‘ Activity 1: Compute weights & balance
**Step-by-step check:**
1. PS Model: `smf.logit("A ~ L1 + L2 + L3", data=df).fit()`
2. Weights: `np.where(A==1, 1/ps, 1/(1-ps))`.
3. Overlap Check: Ensure both curves cover most of the 0-1 range.
4. Love Plot: All points for 'Weighted' should be close to 0 (typically < 0.1).

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from phs564_ci.datasets import load_data

df = load_data("l11_iptw.csv")
ps_model = smf.logit("A ~ L1 + L2 + L3", data=df).fit(disp=0)
df['ps'] = ps_model.predict(df)
df['weights'] = np.where(df['A'] == 1, 1/df['ps'], 1/(1-df['ps']))

weighted_model = smf.wls("Y ~ A", data=df, weights=df['weights']).fit(cov_type='HC1')
print(f"Adjusted ATE: {weighted_model.params['A']:.3f}")
```
