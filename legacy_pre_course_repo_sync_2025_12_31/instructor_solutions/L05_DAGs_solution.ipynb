## Backdoor Path Drill
**DAG:**
- `L1 -> A`
- `L1 -> L2`
- `L2 -> Y`
- `A -> Y`

**Backdoor Path:** `A <- L1 -> L2 -> Y`.
**Adjustment Set:** Either `{L1}` or `{L2}` or both. `{L1}` is preferred as it occurs before treatment and the mediator $L2$.
